```python
import numpy as np
import csv
import codecs
import io
final = []
ahp_matrix = np.matrix([
    [1, 1/3, 4/3, 2/3],
    [3, 1, 4, 2],
    [3/4, 1/4, 1, 1/2],
    [3/2, 1/2, 2, 1]
])
file_name = "ahp.csv"
with io.open('./ahp.csv', 'r', encoding="utf_8_sig")  as f:
    reader = csv.reader(f, quoting=csv.QUOTE_NONNUMERIC)
    for row in reader:
        row = np.matrix(row)
        result = np.dot(ahp_matrix, row.T)
        
        answer = 0        
        result = np.array(result).flatten()

        for j in range(len(result)):
                if j == 0 :
                    answer += result[j]*9
                elif j == 1:
                    answer += result[j]
                elif j == 2:
                    answer += result[j]*5
                else:
                    answer += result[j]*3
        final.append(answer)
        
print(final)
print (max(final))

for i in range(len(final)):
    if final[i]== max(final):
        print(i +1)
                    
```
